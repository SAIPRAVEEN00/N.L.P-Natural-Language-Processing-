```python
# nltk ---> Natural language toolkit
# Spacy
# Gensim
# StanfordNLP

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd 

documents=['Hyderabad is famous for Biryani.I love to stay here',
           'I live in Hyderabad.I would recommend Hyderabad anyone.',
           'I started to like Hyderabad for its Biryani'
           'The next IPL match is in Hyderabad.I enjoy watching IPL here in Hyderabad'
          ]

cv=CountVectorizer()
cv.fit_transform(documents)
```

```python
dt_matrix=cv.fit_transform(documents)

dt_matrix.toarray()
```

```python
cv.get_feature_names()
```

```python
df=pd.DataFrame(dt_matrix.toarray(),columns=cv.get_feature_names())

df
```

```python
tfidf=TfidfVectorizer()

tf_idf_dtm=tfidf.fit_transform(documents)

tf_idf_dtm.toarray()
```
